```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Sales Data.csv',encoding='unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop([ 'Status' ,'unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'].dtype
```

```python
df['Amount']=df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
#describe
df[['Age','Orders','Amount']].describe()
```

```python
# amount spend by gender
sales_gen=df.groupby('Gender')['Amount'].sum().reset_index()

palettle_col={'M':'#e272d5','F':'#e68193'}
plt.figure(figsize=(10,4))
bx=sns.barplot(x='Gender',y='Amount', data=sales_gen,hue='Gender',palette=palettle_col)
plt.title('Male vs. Female Amount Spending Statistics',fontsize=16,fontweight='bold')
# woman puchsing more then man 
```

```python
order_frequency = df.groupby('User_ID')['Orders'].count()
repeat_customers = order_frequency[order_frequency > 1]

repeat_customers_data = df[df['User_ID'].isin(repeat_customers.index)]
repeat_customers_contribution = repeat_customers_data.groupby('User_ID')['Amount'].sum()

plt.figure(figsize=(12,5))
sns.histplot(order_frequency, bins=20, kde=True, color='#ee5f8d')
plt.title('Distribution of Order Frequency per Customer',fontsize=16,fontweight='bold')
plt.xlabel('Number of Orders')
plt.ylabel('Frequency')
plt.show()
```

```python
# genderwise Frequency of Shopping

palette_color={'F':'#e272d5', 'M':'#e47bb5'}
plt.figure(figsize=(10,4))
gender_Wise=sns.countplot(x='Gender',data=df,palette=palette_color,hue='Gender')
plt.title('Frequency of Shopping',fontsize=16,fontweight='bold')
```

```python
# shoping by age group 

p_col={'M':'#e272d5','F':'#f1718d'}
plt.figure(figsize=(10,4))
age_group=sns.countplot(x='Age Group',data=df,hue='Gender',palette=p_col)
for i in age_group.containers:
    age_group.bar_label(i)
plt.title('Age Distribution of Shopping Categories',fontsize=16,fontweight='bold')
```

```python
# amount spand by age group

amt_Age_group=df.groupby('Age Group')['Amount'].sum().reset_index().sort_values(by=('Amount'),ascending=True)
plt.figure(figsize=(10,4))
amt_Age_group_bar=sns.barplot(x='Age Group',y='Amount',data=amt_Age_group,hue='Age Group')
plt.title('Age Group Expenditure',fontsize=16,fontweight='bold')
```

```python
# Top 10 Indian States By Order

order_by_state=df.groupby('State')['Orders'].sum().reset_index().sort_values(by=('Orders'),ascending=False).head(11)
plt.figure(figsize=(10,4))
order_by_state_plot=sns.barplot(x='State',y='Orders',data=order_by_state,hue='State')
for i in order_by_state_plot.containers:
    order_by_state_plot.bar_label(i)
plt.title('Top 10 Indian States By Order',fontsize=16,fontweight='bold')
order_by_state_plot.tick_params(axis='x',labelsize=5.5)
```

```python
# order by Occupation 
o_occ=df.groupby('Occupation')['Orders'].sum().reset_index().sort_values(by=('Orders'),ascending=False)
plt.figure(figsize=(10,4))
o_occ_bar=sns.barplot(x='Orders',y='Occupation',data=o_occ ,hue='Occupation')
for i in o_occ_bar.containers:
    o_occ_bar.bar_label(i)
plt.title('Order By Occupassion',fontsize=16,fontweight='bold')
```

```python
# most popular product by amount
product_amount=df.groupby('Product_Category')['Amount'].sum().reset_index().sort_values(by=('Amount'),ascending=False).head(11)
plt.figure(figsize=(13,4))
product_amt_bar=sns.barplot(x='Product_Category',y='Amount',data=product_amount,hue='Product_Category')
plt.title('Most Popular Products',fontsize=16,fontweight='bold')
product_amt_bar.tick_params(axis='both', labelsize=6)
```

```python
# Cust_name	,amount
top_customer=df.groupby('Cust_name')['Amount'].sum().reset_index().sort_values(by=('Amount'),ascending=False).head(11)
plt.figure(figsize=(12,4.5))
top_customer_plot=sns.barplot(x='Amount',y='Cust_name',data=top_customer,hue='Cust_name')
for i in top_customer_plot.containers:
    top_customer_plot.bar_label(i)
plt.title('Top 10 Customers',fontsize=16,fontweight='bold')
```

```python
# order  by regien 
order_by_zone=df.groupby('Zone')['Orders'].sum().reset_index().sort_values(by=('Orders'),ascending=True)
plt.figure(figsize=(10,4))
order_by_zone_plot=sns.barplot(x='Zone',y='Orders',data=order_by_zone,hue='Zone')
plt.title( 'Regional Shopping Analysis',fontsize=16,fontweight='bold')
 
```
